```python
# get block-level shapefiles for all 50 states

import requests
import zipfile
import os
from pathlib import Path

url1 = 'https://www2.census.gov/geo/tiger/TIGER2020PL/STATE/'
filepath = 'C://districting-data-2020//'

for state in us_state_to_abbrev.keys():
    
    state_abbrev = us_state_to_abbrev[state]
    state_code = state_codes[state_abbrev]
    
    # download the zip file
    url2 = state_code + '_' + state.upper() + '/' + state_code + '/' 
    webfilename = 'tl_2020_' + state_code + '_tabblock20.zip'
    webfile = requests.get( url1 + url2 + webfilename )
    open(filepath+state_abbrev+'_block.zip', 'wb').write(webfile.content)
    
    # unzip it
    zipfilename = state_abbrev + '_block.zip'
    with zipfile.ZipFile(filepath+zipfilename,"r") as zip_ref:
        zip_ref.extractall(filepath)
        
    # rename the unzipped files
    oldfilename = 'tl_2020_' + state_code + '_tabblock20.'
    newfilename = state_abbrev + "_block."
    for fx in {'shp','prj','dbf','cpg','shx'}:
        path = Path( filepath + oldfilename + fx )
        if path.is_file():
            os.rename(filepath + oldfilename + fx, filepath + newfilename + fx)
        else:
            print("file not available for",state)
```
